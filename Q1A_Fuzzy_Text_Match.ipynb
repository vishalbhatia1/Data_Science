```python
# requires fuzzywuzzy and python-Levenshtein
# fuzzywuzzy uses Levenshtein Distance to calculate the differences between sequences
import pandas as pd
from fuzzywuzzy import process 
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_excel(r'E:\DocumentsData.xlsx')
```

```python
data = df['Doc Name'].astype(str).to_list()
```

```python
x_train, x_test = train_test_split(data, test_size=0.05, random_state=42)
```

```python
print("Train size: ", len(x_train), "\nTest size: ", len(x_test))
```

```python
def get_matches(query, choices, limit=1):
    print(query, choices, limit)
    results =  process.extract(query, choices, limit)
    return results
```

```python
columns = ['Test String', 'Count of Matches', 'Highest Confidence', 'Most Confident String']
test_res_df = pd.DataFrame(columns=columns)

for i in x_test:
    res = process.extract(i, x_train)   
    res_dict = {
        'Test String': i,
        'Count of Matches': len(res),
        'Highest Confidence': res[0][1],
        'Most Confident String': res[0][0]
    }
    test_res_df = test_res_df.append(res_dict, ignore_index=True)
```
